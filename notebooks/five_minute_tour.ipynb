# PydanTable — five-minute tour

Mirror of [docs/QUICKSTART.md](https://pydantable.readthedocs.io/en/latest/QUICKSTART.html). Requires `pydantable` with the Rust extension built.

```python
from pydantic import BaseModel
from pydantable import DataFrame


class Row(BaseModel):
    id: int
    score: float
    label: str


df = DataFrame[Row](
    {
        "id": [1, 2, 3],
        "score": [10.0, 20.5, 7.0],
        "label": ["a", "b", "a"],
    }
)
df
```

```python
df.columns, df.shape, df.info()
```

```python
print(df.describe())
```

```python
filtered = df.filter(df.score > 8.0)
filtered.to_dict()
```
